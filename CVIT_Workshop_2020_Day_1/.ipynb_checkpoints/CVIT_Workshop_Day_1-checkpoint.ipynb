```python
# Define our figure. Figsize is given in inches
fig = plt.figure(figsize=(16, 16))

# compure the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
# set up a subplot grid that has height 2 and widht 1, 
# and set the first such subplot as active
ax = fig.add_subplot(221)
# make the first plot
ax.plot(x, y_sin)
ax.set_title('Sine')

# set the second subplot as active, and make the second plot.
# @param: c = color of the plot
# @param: linewidth = width of the line
ax = fig.add_subplot(222)
ax.plot(x, y_cos, linewidth=2, c='orange')
ax.set_title('Cosine')

# third subplot plot + scatter
# marker is the symbol used for scatter plot.
ax = fig.add_subplot(223)
ax.plot(x, y_sin/y_cos)
ax.scatter(x, y_sin/y_cos, c='red', marker='s')
ax.set_title('Tangent')

# Co Tangent
ax = fig.add_subplot(224)
ax.plot(x, y_cos/y_sin)
ax.scatter(x, y_cos/y_sin, c='green', marker='o')
ax.set_title('Cotangent')

# remove axis
ax.axis("off")
```

```python
N = 5
A = (20, 35, 30, 35, 27)
B = (25, 32, 34, 20, 25)
A_std = (2, 3, 4, 1, 2)
B_std = (3, 5, 2, 3, 3)
ind = np.arange(N) # the x locations for the groups

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=ind, 
    y=A, 
    error_y=dict(
        type='data', # value of error bar given in data coordinates
        array=A_std,
        visible=True,
        color='green'
        ),
        name='A'
    ))
fig.add_trace(go.Scatter(
    x=ind,
    y=B,
    error_y=dict(
        type='data',
        array=B_std,
        visible=True,
        color='green'
        ),
        name='B'
))
fig.update_layout(title = 'A & B', xaxis_title = "Sequence length", yaxis_title="Index")
fig.show()
```

We know from the mid school mathematics that the distance from origin to the line  ax+by+c=0  is  |c| / √(a2+b2) .

Equivalently, distance from the origin to the line  wTX+c=0  is given by  |−c| / ||w||

Question: Can you write code to find distance of the line  x=50  and  y=25  from the origin?

Can you derive the formula to compute distance between two parallel lines from the formulas provided in the previous section?

```python
# Task 7

import math

def distance_from_origin(a, b, c):
  d = abs(c) / math.sqrt(a*a + b*b)
  return d

def distance_between_pll_lines(a1, b1, c1, a2, b2, c2):
  if c1 * c2 < 0:
    return distance_from_origin(a1, b1, c1) + distance_from_origin(a2, b2, c2)
  else:
    return abs(distance_from_origin(a1, b1, c1) - distance_from_origin(a2, b2, c2))
```

```python
print(distance_from_origin(1, 0, -50))
print(distance_from_origin(0, 1, -25))
print(distance_between_pll_lines(1, 0, -50, 1, 0, -20))
```

```python
# Task 9

# square plot
a = 10
x = np.linspace(-a/2, a/2, 200)
y = np.linspace(-a/2, a/2, 200)
y1 = (- 0 * x + a/2) / 1.0
y2 = (- 0 * x - a/2) / 1.0
x1 = (- 0 * y + a/2) / 1.0
x2 = (- 0 * y - a/2) / 1.0
%matplotlib inline
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x1, y)
plt.plot(x2, y)
```

```python
# rectangle plot
a = 20
b = 10
x = np.linspace(-a/2, a/2, 200)
y = np.linspace(-b/2, b/2, 200)
y1 = (- 0 * x + b/2) / 1.0
y2 = (- 0 * x - b/2) / 1.0
x1 = (- 0 * y + a/2) / 1.0
x2 = (- 0 * y - a/2) / 1.0
%matplotlib inline
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x1, y)
plt.plot(x2, y)
```

```python
# eclipse plot
a = 5
b = 2
x = np.linspace(-a, a, 200)
y_plus = b * np.sqrt(1 - (x**2) / (a**2))
y_minus = -b * np.sqrt(1 - (x**2) / (a**2))
%matplotlib inline
plt.plot(x, y_plus)
plt.plot(x, y_minus)
```

```python
from mpl_toolkits.mplot3d import Axes3D

a = 1
b = 1
c = 0
d = 10

xx = np.linspace(-10, 10, 20)
mean_x = np.mean(xx)
var_x = np.var(xx)
zz = 1/(var_x * np.sqrt(2*np.pi)) * np.exp(-((xx-mean_x)**2) / (2*(var_x**2)))
X, Z = np.meshgrid(xx, zz)
yy = (-a*X - c*Z - d) * 1.0 / b
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, yy, Z)
```

```python
from mpl_toolkits.mplot3d import Axes3D

xx = np.linspace(-10, 10, 20)
mean_x = np.mean(xx)
var_x = np.var(xx)
yy = np.linspace(-10, 10, 20)
mean_y = np.mean(yy)
var_y = np.var(yy)
X, Y = np.meshgrid(xx, yy)
zz = 1 / (2 * np.pi * var_x * var_y) * np.exp(-(((X-mean_x)**2)/(2*(var_x**2)) + ((Y-mean_y)**2)/(2*(var_y**2))))
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, Y, zz)
```
